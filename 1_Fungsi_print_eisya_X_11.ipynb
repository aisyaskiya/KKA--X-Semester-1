#Fungsi Parameter

```python
# Latihan
# Silakan tampilkan keluaran/ouput berupa:
# Haidar*Adzka*Daffanes*Adzka
```

```python
print("Nama_siswa_1")
```

```python
nama_siswa_1 = "Daffanes"
nama_siswa_2 = "Adzka"
nama_siswa_3 = "Haidar"

print(nama_siswa_1, nama_siswa_2, sep="-")
```

```python
print(nama_siswa_1)
print(nama_siswa_2)
```

```python
print(nama_siswa_1, nama_siswa_2)
```

```python
print(nama_siswa_1, end="-")
print(nama_siswa_2)
```

Latihan soal:

Gunakan parameter end untuk menampilka tulisan sebagai berikut:
Haidar#Daffanes#Adzka

```python
print(nama_siswa_1, end="-")
print(nama_siswa_2, end="-")
print(nama_siswa_3)
```
